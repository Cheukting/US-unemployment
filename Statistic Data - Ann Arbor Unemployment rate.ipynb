```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
us_data = pd.read_csv('US Unemployment rate.csv')
us_data.head()
```

```python
general = us_data.loc[us_data.Location=='United States, General']
general.head()
```

```python
high = us_data.loc[us_data.Location=='United States, Highest State']
high.head()
```

```python
low = us_data.loc[us_data.Location=='United States, Lowest State']
low.head()
```

```python
aa_data = pd.read_csv('Ann Arbor Unemployment rate.csv')
aa_data.head()
```

```python
%matplotlib notebook
plt.plot(aa_data.Year,aa_data['Unemployment Rate'],color='red')
plt.plot(general.Year,general['Unemployment Rate'],color='black')
```

```python
plt.gca().fill_between(general.Year, 
                       low['Unemployment Rate'], high['Unemployment Rate'], 
                       facecolor='grey', 
                       alpha=0.25)
```

```python
plt.gca().set_xticks(general.Year)
```

```python
plt.title('Unemployment rate of Ann Arbor, Michigan, US in 2013-2017')
```

```python
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```

```python
style = dict(size=8, color='black')

us_2017 = general['Unemployment Rate'].iloc[0]
plt.gca().text(2017, us_2017, 'US: {}'.format(us_2017),ha='left', va='center', **style)
```

```python
style = dict(size=8, color='red')

aa_2017 = aa_data['Unemployment Rate'].iloc[0]
plt.gca().text(2017, aa_2017, 'Ann Arbor: {}'.format(aa_2017),ha='left', va='center', **style)
```

```python
style = dict(size=8, color='grey')

low_2017 = low['Unemployment Rate'].iloc[0]
plt.gca().text(2017, low_2017, 'lowest state',ha='left', va='center', **style)
high_2017 = high['Unemployment Rate'].iloc[0]
plt.gca().text(2017, high_2017, 'highest state',ha='left', va='center', **style)
```
